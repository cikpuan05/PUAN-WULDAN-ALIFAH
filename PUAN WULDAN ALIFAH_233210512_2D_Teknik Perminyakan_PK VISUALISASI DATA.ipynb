```python
import matplotlib.pyplot as plt

usia = ['20', '21', '22', '23', '24']
gender = [4, 4, 3, 3, 3 ]
#grafiknya
plt.figure(figsize=(8, 4))
plt.plot(usia, gender, marker='*', color='g', linestyle='-')
plt.title('usia berdasarkan gender')
plt.xlabel('usia')
plt.grid(True)
plt.show()
```

## line chart ini berisi usia berdasarkan gender adalah 20 = 4 orang, 21 = 4 orang, 22 = 3 orang, 23 = 3 orang, 24 = 3 orang. Setelah coding Bar chart dibuat, didapatkan Bar chart seperti dibawah ini.

```python
import matplotlib.pyplot as plt

# Data contoh
usia = ['20', '21', '22', '23', '24']
gender = [4, 4, 3, 3, 3 ]

#grafiknya
plt.figure(figsize=(4, 4))
plt.pie(gender ,labels=usia, autopct="%1.1f%%", startangle=140)
plt.title('Pie Chart gender per usia')
plt.axis('equal') # Agar Lingkaran berbentuk Lingkaran sempurna
plt.show()
```

# Bar chart ini berisi usia berdasarkan gender, usia berdasarkan gender adalah 20 = 4 orang, 21 = 4 orang, 22 = 3 orang, 23 = 3 orang, 24 = 3 orang. Setelah coding Bar chart dibuat, didapatkan Bar chart seperti dibawah ini.

```python

import matplotlib.pyplot as plt

usia = ['20', '21', '22', '23', '24']
gender = [4, 4, 3, 3, 3 ]

#grafiknya
plt.figure(figsize=(8, 4))
plt.bar(usia, gender, color='b') 
plt.title('Line usia berdasarkan gender ')
plt.xlabel('usia')
plt.ylabel('gender')
plt.grid(False)
plt.show()
```
